# Preparing the data

```python
# Importing the required libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 
```

```python
# Read in the data

df = pd.read_csv(r'C:\Users\T.SHARMISHTA\OneDrive\Desktop\Job Search\Python - Movie Data Exploration\movies.csv')
```

```python
# GLimpsing into the data

df.head()
```

# Cleaning the data

```python
# Check for missing data

df.isnull().sum()
```

```python
# Dropping the columns with null values

df = df.dropna()
```

```python
# Checking te data types

types = df.dtypes
print(types)
```

```python
# Changing the datatypes 

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

# Extracting the year

df['yearnew'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Sorting the data based on the gross revenue

df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Dropping the duplicates

df.drop_duplicates()
```

# Finding Correlations in the data

```python
# Looking at any two features - Budget vs Gross Revenue

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
# Correlation Matrix

corr_matrix = df.corr(numeric_only = True)
sns.heatmap(corr_matrix, annot = True)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
# Looking at non-numeric features

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
# Correlation Matrix for the numerized data

corr_matrix = df_numerized.corr(numeric_only = True)
sns.heatmap(corr_matrix, annot = True)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
corr_mat = df_numerized.corr()
corr_pairs = corr_mat.unstack().sort_values()
corr_pairs
```

```python
high_corr = corr_pairs[(corr_pairs) > 0.5]
high_corr
```

We can conclude that votes and budget have the highest correlation to the gross revenue.
